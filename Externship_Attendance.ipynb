#### Creating Visualizations

```python
newData = newData.drop(columns = ['Points', 'Completed?'])
newData
#traverse through columns and count values in each column, put it into series and T old table and insert that series as new column
```

```python
Attendance = [np.count_nonzero(np.array(newData.get(newData.columns[i]))) for i in range(newData.shape[1])]
Total = [249 for i in range(newData.shape[1])]
newData = newData.T
index_names = np.array(newData.reset_index().get('index'))
```

```python
chrono_df = pd.DataFrame({'Event Attendance': Attendance, 'Total Participants': Total}, index = index_names)
chrono_df['Percentage'] = [i / j * 100 for i,j in zip(chrono_df['Event Attendance'], chrono_df['Total Participants'])]
chrono_df = chrono_df.reset_index().rename(columns={"index": "Event Name"})
chrono_df
```

```python
desc_df = chrono_df.sort_values('Event Attendance', ascending=False)
desc_df
```

```python
plt.figure(figsize=(15,10))
bar1 = sns.barplot(x="Event Name", y="Total Participants", data=chrono_df, color='lightblue')
bar2 = sns.barplot(x="Event Name",  y="Event Attendance", data=chrono_df, color='darkblue')

top_bar = mpatches.Patch(color='darkblue', label='Attendance')
bottom_bar = mpatches.Patch(color='lightblue', label='Total Participants')
plt.legend(handles=[top_bar, bottom_bar])
plt.title("Percentage Attendance of Each Event by Chronological Order", fontsize = 20)

locs, labels = plt.xticks()
plt.setp(labels, rotation=30)

plt.show()
```

```python
plt.figure(figsize=(15,10))
bar1 = sns.barplot(x="Event Name", y="Total Participants", data=desc_df, color='lightblue')
bar2 = sns.barplot(x="Event Name",  y="Event Attendance", data=desc_df, color='darkblue')

top_bar = mpatches.Patch(color='darkblue', label='Attendance')
bottom_bar = mpatches.Patch(color='lightblue', label='Total Participants')
plt.legend(handles=[top_bar, bottom_bar])
plt.title("Percentage Attendance of Each Event by Descending Order", fontsize = 20)

locs, labels = plt.xticks()
plt.setp(labels, rotation=30)

plt.show()
```
